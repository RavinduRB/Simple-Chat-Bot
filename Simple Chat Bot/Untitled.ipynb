```python
import nltk
import random
import string
import tkinter as tk
from tkinter import *

# Initialize NLTK resources (download if necessary)
nltk.download('punkt')
nltk.download('wordnet')

# Sample text data for the chatbot
raw = """Hello! How can I help you today? Chatbots are artificial intelligence systems that interact with users via text or voice interfaces. They can answer questions, provide recommendations, and assist with various tasks."""

# Convert to lowercase
raw = raw.lower()

# Tokenization
sent_tokens = nltk.sent_tokenize(raw)  # converts to list of sentences
word_tokens = nltk.word_tokenize(raw)  # converts to list of words

# Preprocessing
lemmer = nltk.stem.WordNetLemmatizer()

def LemTokens(tokens):
    return [lemmer.lemmatize(token) for token in tokens]

remove_punct_dict = dict((ord(punct), None) for punct in string.punctuation)

def LemNormalize(text):
    return LemTokens(nltk.word_tokenize(text.lower().translate(remove_punct_dict)))

# Keyword Matching
GREETING_INPUTS = ("hello", "hi", "greetings", "hey")
GREETING_RESPONSES = ["Hi!", "Hey!", "Hi there!", "Hello!", "I am glad you are talking to me!", "Good morning!", "Good afternoon!", "Good night!"]
LOVING_RESPONSES = [
    "You're amazing!", "You are loved!", "Sending you positive vibes!", "You are wonderful!",
    "Keep smiling!", "You brighten my day!", "You are a beautiful person inside and out!",
    "Stay positive!", "You are doing great!", "I'm here for you!"
]
GIRLFRIEND_RESPONSES = [
    "I love you more than words can say.", "You mean the world to me.", "I'm so grateful to have you in my life.",
    "You make my heart skip a beat.", "Thinking of you makes my day brighter.", "You are my everything.",
    "I can't imagine my life without you.", "You are my rock and my best friend.", "I cherish every moment with you.",
    "You make me a better person.", "I'm proud to call you mine.", "Your love fills my heart with joy."
]

def greeting(sentence):
    for word in sentence.split():
        if word.lower() in GREETING_INPUTS:
            return random.choice(GREETING_RESPONSES)
    return None

def loving_words():
    return random.choice(LOVING_RESPONSES)

def girlfriend_words():
    return random.choice(GIRLFRIEND_RESPONSES)

# Generating response
def response(user_response):
    normalized_user_response = LemNormalize(user_response)
    bot_response = ''
    if greeting(' '.join(normalized_user_response)) is None:
        if 'i love you' in ' '.join(normalized_user_response):
            bot_response = "I love you too!"
        elif 'what are you doing now' in ' '.join(normalized_user_response):
            bot_response = "I'm looking at you"
        elif 'i miss you' in ' '.join(normalized_user_response):
            bot_response = "I miss you too but I have no way to come to you."
        elif 'bye babe' in ' '.join(normalized_user_response):
            bot_response = "Bye ‚ù§. Be careful. I miss you always"
        else:
            bot_response = "What are you saying my dear. I don't understand. " + loving_words() + " " + girlfriend_words()
    else:
        bot_response = greeting(' '.join(normalized_user_response)) + " " + loving_words() + " " + girlfriend_words()
    return bot_response

# Tkinter GUI
def send():
    global word_tokens, sent_tokens

    user_response = entry_box.get("1.0", "end-1c").strip().lower()
    entry_box.delete("0.0", END)
    
    if user_response != 'bye':
        if user_response in ('thanks', 'thank you'):
            chat_log.config(state=NORMAL)
            chat_log.insert(END, "You: " + user_response + '\n\n')
            chat_log.insert(END, "Lucy: You are welcome! " + loving_words() + " " + girlfriend_words() + "\n\n")
            chat_log.config(state=DISABLED)
            chat_log.yview(END)
        else:
            if greeting(user_response) is not None:
                chat_log.config(state=NORMAL)
                chat_log.insert(END, "You: " + user_response + '\n\n')
                chat_log.insert(END, "Lucy: " + greeting(user_response) + " " + loving_words() + " " + girlfriend_words() + '\n\n')
                chat_log.config(state=DISABLED)
                chat_log.yview(END)
            else:
                sent_tokens.append(user_response)
                word_tokens += nltk.word_tokenize(user_response)
                final_words = list(set(word_tokens))
                chat_log.config(state=NORMAL)
                chat_log.insert(END, "You: " + user_response + '\n\n')
                chat_log.insert(END, "Lucy: " + response(user_response) + '\n\n')
                chat_log.config(state=DISABLED)
                chat_log.yview(END)
                sent_tokens.remove(user_response)
    else:
        chat_log.config(state=NORMAL)
        chat_log.insert(END, "You: " + user_response + '\n\n')
        chat_log.insert(END, "Lucy: Bye! Take care! " + loving_words() + " " + girlfriend_words() + "\n\n")
        chat_log.config(state=DISABLED)
        chat_log.yview(END)
        window.quit()

# Create tkinter window
window = Tk()
window.title("Princess Lucy")
window.geometry("400x500")
window.resizable(width=FALSE, height=FALSE)

# Create Chat window
chat_log = Text(window, bd=0, bg="white", height="8", width="50", font="Arial",)
chat_log.config(state=DISABLED)

# Bind scrollbar to Chat window
scrollbar = Scrollbar(window, command=chat_log.yview, cursor="heart")
chat_log['yscrollcommand'] = scrollbar.set

# Create a Button to send a message
send_button = Button(window, font=("Verdana",12,'bold'), text="Send", width="12", height=5,
                    bd=0, bg="#32de97", activebackground="#3c9d9b",fg='#ffffff',
                    command=send)

# Create the box to enter the message
entry_box = Text(window, bd=0, bg="white",width="29", height="5", font="Arial")
entry_box.bind("<Return>", lambda event: send())

# Place all components on the screen
scrollbar.place(x=376,y=6, height=386)
chat_log.place(x=6,y=6, height=386, width=370)
entry_box.place(x=6, y=401, height=90, width=265)
send_button.place(x=276, y=401, height=90)

window.mainloop()
```
